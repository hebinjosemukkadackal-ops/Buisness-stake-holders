Summary of Findings (Titanic EDA):
Survival was strongly influenced by passenger gender and class.
Female passengers had significantly higher survival rates compared to males.
Passengers in 1st class were more likely to survive than those in 2nd and 3rd class.
Higher ticket fare showed positive association with survival probability.
Age had weaker correlation with survival, though younger passengers showed slightly better survival.
Most passengers belonged to 3rd class, indicating class imbalance in the dataset.
Missing values were observed in Age and Cabin features, which may affect analysis.
